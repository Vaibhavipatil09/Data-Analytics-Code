```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
c='g'
plt.plot(x,y,c)
plt.show()
```

```python
x=[3,4,5,8]
y=[5,6,3,2]
plt.plot(x,y,'r',marker='o')
plt.show()
```

```python
x=[3,4,5,8]
y=[5,6,3,2]
plt.plot(x,y,'r')
plt.plot(y,marker='>')
plt.plot(x,'g',marker='o')
plt.show()
```

```python
#Bar Graph
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
#Bar Graph
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','y','b']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)
```

```python
#Bar Graph
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','y','b']
plt.bar(x,y,color=c,width=0.3)
plt.show()
```

```python
#Bar Graph
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','y','b']
plt.barh(x,y,color=c,height=0.1)
plt.show()
```

```python
#scatterplot
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=['red','green','magenta','orange','blue','brown']
plt.scatter(x,y,color=c,s=150)
plt.title("Scatter Plot")
plt.xlabel("Months")
plt.ylabel("Numbers")
plt.show()
```
