```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fname= r'inn.jpg'
image=Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image,cmap='gray')
plt.show()
```

```python
fname= r'inn.jpg'
image=Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image,cmap='RdYlGn_r')
plt.show()
```

```python
#piechart
x=[10,20,30,40]
y=["English","Hindi","Maths","Sciece"]
c=["Yellow","Magenta","green","red"]
plt.pie(x,labels=y,colors=c)
plt.title("Subjects")
plt.legend()
plt.savefig("pie.jpg")
plt.show()
```
