```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8]
y=[8,7,6,5,4,3,2,1]
```

```python
#using matplotlib
plt.plot(x,y,color='r')
plt.show()
```

```python
sns.lineplot(x=x,y=y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[8,7,6,5,4,3,2,1]

import pandas as pd

df=pd.DataFrame({"var":x,"var1":y})
```

```python
df
```

```python
sns.lineplot(x='var',y='var1',data=df)
plt.show()
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex')
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex',palette='rocket_r')
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex',style='sex',
             palette='rocket_r',markers=["o",">"])
plt.show()
```

```python
df1=sns.load_dataset('penguins').head(20)
```

```python
df1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex',style='sex',
             palette='rocket_r',markers=["o",">"])
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex',style='sex',
             palette='rocket_r',markers=["o",">"])
plt.grid()
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue='sex',style='sex',
             palette='rocket_r',markers=["o",">"])
plt.grid()
plt.title("LINE GRAPH")
plt.show()
```

```python
sns.histplot(x="bill_length_mm",data=df1)

plt.show()
```
