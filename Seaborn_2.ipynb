```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1["island"]
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1)
plt.show()
```

```python
df1
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex')
plt.show()
```

```python
#order parameter

order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',order=order1)
plt.show()
```

```python

order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',order=order1,
           hue_order=['Female','Male'])
plt.show()
```

```python

order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',order=order1,
           hue_order=['Female','Male'],palette='dark:red',saturation=0.6)
plt.show()
```

```python
#histogram
sns.displot(df1["flipper_length_mm"])
plt.show()
```

```python
#bins parameter for sequencing
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
#color
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],color='red')
plt.show()
```

```python
#kde=kernel density
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,
            color='g')
plt.show()
```
